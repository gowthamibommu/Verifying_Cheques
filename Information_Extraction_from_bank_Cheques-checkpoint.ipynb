```python
import pytesseract
img = cv2.imread("SBI_cheque.jpg",cv2.IMREAD_UNCHANGED)
gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
adaptive_threshold= cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,85,11)
cv2_imshow(adaptive_threshold)
cv2.waitKey(0)
text= pytesseract.image_to_string(adaptive_threshold)
print(text)
```

```python
## Detecting Charactor
import cv2
himg,wimg = adaptive_threshold.shape
imgbox=pytesseract.image_to_boxes(adaptive_threshold)
for b in imgbox.splitlines():
#     print(b)
    b=b.split(' ')
    print(b)
    x,y,w,h = int(b[1]),int(b[2]),int(b[3]),int(b[4])
    cv2.rectangle(adaptive_threshold,(x,himg-y),(w,himg-h),(0,0,255),1)
cv2_imshow(adaptive_threshold)
cv2.waitKey(0)
```
